```python
#Store registration information in the database with active/inactive status
def registration(username, email, password, first_name, last_name, phone_number, account_status="Active"):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("INSERT INTO users (username, email, password, first_name, last_name, phone_number, account_status) VALUES (%s, %s, %s, %s, %s, %s, %s)", (username, email, password, first_name, last_name, phone_number, account_status))
    conn.close()
    return "User Registered Successfully"
```

```python
#Validate Login Credentials
def validate_login(username, password):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("SELECT * FROM users WHERE username = %s AND password = %s", (username, password))
        user = cursor.fetchone()
    conn.close()
    if user:
        return "Login Successful"
    else:
        return "Invalid Credentials"
```

```python
#Retrieve Forgot My Password Information

#Update Forgot My Password
def forgot_password_update(new_password):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("UPDATE users SET password = %s" (new_password))
    conn.close()
    return "Password Updated Successfully"
```

```python
#Retrieve Edit Profile information for user table (exclude Email)
def get_profile_main(user_id):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("SELECT username, password, first_name, last_name, phone_number FROM users WHERE user_id = %s", (user_id,))
        profile = cursor.fetchone()
    conn.close()
    return profile
#Retrieve Edit Profile information for payment table
def get_profile_payment(user_id):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("SELECT card_number, expiration_date, billing_house, billing_street, billing_apt, billing_zip FROM PaymentCard WHERE user_id = %s", (user_id,))
        profile = cursor.fetchone()
    conn.close()
    return profile
#Retrieve Edit Profile information for address
def get_profile_address(user_id):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("SELECT house_number, street, apt, zip FROM MailingAddress WHERE user_id = %s", (user_id,))
        profile = cursor.fetchone()
    conn.close()
    return profile

#Update Edit Profile information (exclude Email)
def update_profile_main(user_id, new_username, new_password, new_first_name, new_last_name, new_phone_number):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("UPDATE users SET username = %s, password = %s, first_name = %s, last_name = %s, phone_number = %s WHERE user_id = %s", (new_username, new_password, new_first_name, new_last_name, new_phone_number, user_id))
    conn.close()
    return "Profile Updated Successfully"
def update_profile_payment(user_id, new_card_number, new_expiration_date, new_billing_house, new_billing_street, new_billing_apt, new_billing_zip):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("UPDATE PaymentCard SET card_number = %s, expiration_date = %s, billing_house = %s, billing_street = %s, billing_apt = %s, billing_zip = %s WHERE user_id = %s", (new_card_number, new_expiration_date, new_billing_house, new_billing_street, new_billing_apt, new_billing_zip, user_id))
    conn.close()
    return "Payment Information Updated Successfully"
def update_profile_address(user_id, new_house_number, new_street, new_apt, new_zip):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("UPDATE MailingAddress SET house_number = %s, street = %s, apt = %s, zip = %s WHERE user_id = %s", (new_house_number, new_street, new_apt, new_zip, user_id))
    conn.close()
    return "Address Updated Successfully"
```

```python
#Retrieve favorites movies
def get_favorites(user_id):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("SELECT movie_id FROM favorites WHERE user_id = %s", (user_id,))
        favorites = cursor.fetchall()
    conn.close()
    return favorites
#Add a movie to favorites
def add_favorite(user_id, movie_id):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("INSERT INTO favorites (user_id, movie_id) VALUES (%s, %s)", (user_id, movie_id))
    conn.close()
    return "Movie Added to Favorites"
```
