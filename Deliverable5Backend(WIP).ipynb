```python
import pymysql

def get_db():
    return pymysql.connect(
        host="127.0.0.1",
        port=33306,
        user="root",
        password="mysqlpass",
        database="cinemaebooking",
        cursorclass=pymysql.cursors.DictCursor
    )
```

```python
#Store registration information in the database with active/inactive status
def registration(username, email, password):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("INSERT INTO users (username, email, password) VALUES (%s, %s, %s)", (username, email, password))
    conn.close()
    return "User Registered Successfully"
```

```python
#Validate Login Credentials
def validate_login(username, password):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("SELECT * FROM users WHERE username = %s AND password = %s", (username, password))
        user = cursor.fetchone()
    conn.close()
    if user:
        return "Login Successful"
    else:
        return "Invalid Credentials"
```

```python
#Update Forgot My Password
def forgot_password_update(new_password):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("UPDATE users SET password = %s" (new_password))
    conn.close()
    return "Password Updated Successfully"
```

```python
#Retrieve Edit Profile information (exclude Email)
def get_profile(username, password):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("SELECT username, password FROM users WHERE username = %s AND password = %s", (username, password))
        profile = cursor.fetchone()
    conn.close()
    return profile

#Update Edit Profile information (exclude Email)
def update_profile(username, password, new_username, new_password):
    conn = get_db()
    with conn.cursor() as cursor:
        cursor.execute("UPDATE users SET username = %s, password = %s WHERE username = %s AND password = %s", (new_username, new_password, username, password))
    conn.close()
    return "Profile Updated Successfully"
```
